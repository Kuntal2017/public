
## Matplotlib

### 1. Line Plot

```python
import numpy as np
from matplotlib import pyplot as plt
#import matplotlib.pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = x * 0.5
y
```

```python
#Linear plot 
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

```python
#changed line color to red and line style to dotted and line width to 3
plt.plot(x,y,color='r',linestyle = ':',linewidth = '3')
plt.title('Line Plot')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

```python
plt.plot(x,y,color='r',linestyle = 'dashed',linewidth = '3',
         marker = 'o',markersize = ' 12',markerfacecolor = 'b'    )
plt.title('Line Plot')
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

```python
# working with multiple lines
x = np.arange(1,11)
y1 = x*2
y2 = x*3
y3 = x*4
```

```python
plt.plot(x,y1,linestyle='dashed',color='b',linewidth='3')
plt.plot(x,y2,linestyle=':',color='r',linewidth='5')
plt.plot(x,y3,linestyle='-.',color='g',linewidth='7')
plt.title('Line plot with multiple lines')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.grid(True)
plt.show()
```

```python
#sub plot method - plt.subplot(rows,column,number)
plt.subplot(1,3,1)
plt.plot(x,y1,linestyle='dashed',color='b',linewidth='3')
plt.xlabel('x-label')
plt.ylabel('y-label')

#second subplot complete
plt.subplot(1,3,2)
plt.plot(x,y2,linestyle=':',color='r',linewidth='5')
plt.xlabel('x-label')
plt.ylabel('y-label')
#third subplot 
plt.subplot(1,3,3)

plt.plot(x,y3,linestyle='-.',color='g',linewidth='7')
plt.title('Line plot with multiple lines')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.grid(True)
plt.show()

#in every subplot we have to mention x and y labels , title, grid
```

### 2. Bar plot or Bar graph (used for categorical data)

```python
#vertical bar chart
student = {'jack':45,'snow':89,'dragon':25}
names = list(student.keys())
marks = list(student.values())
plt.bar(names,marks)
plt.title('class 5 marks')
plt.xlabel('Student Names')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
#horizontal bar chart

plt.barh(names,marks,color='y')
plt.title('class 5 marks')
plt.ylabel('Student Names')
plt.xlabel('Marks')
plt.grid(True)
plt.show()
```

### 3. Scatter Plot

```python
#x =np.arange(10,100,10)
#a = np.random.randint(1,20,10)
x = [10,20,30,40,50,60,70,80,90]
a = [9,3,6,5,2,0,8,1,5]
plt.scatter(x,a,color='r')
plt.title('scatter Plot')
plt.show()
```

```python
x =np.arange(10,100,10) #9 values

a = np.random.randint(1,20,9) # 9 values
# here for size we use s
plt.scatter(x,a,color='r',marker='*', s=100)
plt.title('scatter Plot')
plt.show()
```

```python
#two scatter plots in one graph
x =np.arange(10,100,10) #9 values

a = np.random.randint(1,20,9) # 9 values
b = np.random.randint(1,20,9) # 9 values
# here for size we use s
plt.scatter(x,a,color='r',marker='*', s=100)
plt.scatter(x,b,color='g', marker='.',s=100)
plt.title('scatter Plot')
plt.show()
```

```python
#two scatter plots 
x =np.arange(10,100,10) #9 values

a = np.random.randint(1,20,9) # 9 values
b = np.random.randint(1,20,9) # 9 values
# here for size we use s
plt.subplot(1,2,1)
plt.scatter(x,a,color='r',marker='*', s=100)
plt.title('scatter plot 1')
plt.subplot(1,2,2)
plt.scatter(x,b,color='g', marker='.',s=100)
plt.title('scatter Plot2')
plt.show()
```

### 4.Histogram = used for numerical data

```python
#A histogram displays numerical data by grouping data into "bins" of
#  equal width. Each bin is plotted as a bar whose height corresponds to
#  how many data points are in that bin. 
#Bins are also sometimes called "intervals", "classes", or "buckets"

data = [1,2,3,4,5,5,5,6,2,3,7,8,9,9,5,1,2,3,3,7,18,19]
plt.hist(data,color='c')
plt.show()
```

### 7. Pie chart 

```python
stock = ['nuts','fruits','Dry-fruits','Choclates']
quantity = [90,110,28,10]
plt.pie(quantity,labels=stock)
plt.show()
```

```python
#now  to display percentage
stock = ['nuts','fruits','Dry-fruits','Choclates']
quantity = [90,110,28,10]
# 0.2 means 2 places after decimal
plt.pie(quantity,labels=stock,autopct='%0.2f%%',
        colors=['yellow','red','grey','green'])
plt.show()
```

### 8. Doughtnut chart (similar to pie chart just middle circle is white)

```python
stock = ['nuts','fruits','Dry-fruits','Choclates']
quantity = [90,110,28,10]
plt.pie(quantity,labels=stock,radius=1)
plt.pie([1],colors=['white'],radius=0.5)
plt.show()
```

```python
#now  to display percentage
stock = ['nuts','fruits','Dry-fruits','Choclates']
quantity = [90,110,28,10]
explode = (0.1,0,0,0) #to explode the slice of pie
# 0.2 means 2 places after decimal
plt.pie(quantity,labels=stock,autopct='%0.2f%%',
        colors=['yellow','red','grey','green'],explode=explode,shadow=True)
plt.show()
```
